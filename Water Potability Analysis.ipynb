```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as plx
```

```python
df = pd.read_csv("water_potability.csv")
df.head()
```

# Cleaning the Data and Basic EDA

```python
#checking for null values
df.isnull().sum()
```

```python
#removing null values
df = df.dropna()
df.isnull().sum()
```

```python
df.info()
```

```python
df.describe()
```

```python
df_corr = df.corr()
plt.figure(figsize=(15,6))
sns.heatmap(df_corr, annot=True, fmt=".2f");
```

# Visualization and Analysis

```python
plt.figure(figsize=(15,10))
sns.set_style('whitegrid')
sns.countplot(x='Potability', data=df)
plt.title("Distribution of Undrinkable and Drinkabke water")
```

```python
#chart showing water pH distribution on water potability
f = plt.figure(figsize=(10,7))
ax = f.add_subplot(1,1,1)
sns.histplot(data=df, ax=ax, stat="count", multiple= "stack", x="ph", hue = "Potability", 
             element= "bars", legend=True)
ax.set_title("Factors Affecting Water Quality: pH")
ax.set_xlabel("pH")
ax.set_ylabel("Count")
```
